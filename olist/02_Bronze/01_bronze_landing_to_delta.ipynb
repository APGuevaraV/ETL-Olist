```python
DATASETS = {
    "olist_customers": {
        "file": "olist_customers_dataset.csv",
        "multiline": False
    },
    "olist_orders": {
        "file": "olist_orders_dataset.csv",
        "multiline": False
    },
    "olist_order_items": {
        "file": "olist_order_items_dataset.csv",
        "multiline": False
    },
    "olist_order_payments": {
        "file": "olist_order_payments_dataset.csv",
        "multiline": False
    },
    "olist_order_reviews": {
        "file": "olist_order_reviews_dataset.csv",
        "multiline": True
    },
    "olist_products": {
        "file": "olist_products_dataset.csv",
        "multiline": False
    },
    "olist_sellers": {
        "file": "olist_sellers_dataset.csv",
        "multiline": False
    },
    "olist_geolocation": {
        "file": "olist_geolocation_dataset.csv",
        "multiline": False
    },
    "product_category_name_translation": {
        "file": "product_category_name_translation.csv",
        "multiline": False
    }
}
```

```python
def read_csv(dataset, config):
    reader = (
        spark.read
             .option("header", "true")
             .option("encoding", "UTF-8")
             .option("mode", "PERMISSIVE")
    )
    if config["multiline"]:
        reader = (
            reader
            .option("multiLine", "true")
            .option("quote", '"')
            .option("escape", '"')
        )
    return reader.csv(f"{BASE_LANDING_PATH}/{FOLDER}/" + config["file"])
```

```python
for dataset,config in DATASETS.items():
    df = read_csv(dataset, config)
    (
        df.write
          .mode("overwrite")
          .format("delta").save(f"{BASE_BRONZE_PATH}/{FOLDER}/{dataset}")
    )
```
